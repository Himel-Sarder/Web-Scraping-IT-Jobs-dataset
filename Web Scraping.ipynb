### **1xx: Informational Responses**
These indicate that the server has received the request and is continuing to process it.
- **100 Continue**: The server has received the request headers, and the client can proceed to send the request body.
- **101 Switching Protocols**: The server is switching to a different protocol as requested by the client.
- **102 Processing**: The server has received the request but has not completed processing it yet.

---

### **2xx: Success**
These indicate that the request was successfully received, understood, and accepted.
- **200 OK**: The request was successful, and the server has returned the requested data.
- **201 Created**: The request was successful, and a new resource was created as a result.
- **202 Accepted**: The request has been received but not yet acted upon.
- **203 Non-Authoritative Information**: The server is returning information from a third party, not the original server.
- **204 No Content**: The server successfully processed the request but did not return any content.
- **205 Reset Content**: The server successfully processed the request and asks the client to reset the view.
- **206 Partial Content**: The server is delivering only part of the resource due to a range header sent by the client.

---

### **3xx: Redirection**
These indicate that further action is needed to complete the request.
- **300 Multiple Choices**: The request has more than one possible response; the user or client must choose one.
- **301 Moved Permanently**: The requested resource has been permanently moved to a new URL.
- **302 Found**: The requested resource resides temporarily under a different URL.
- **303 See Other**: The server is redirecting the client to a different URL for the requested resource.
- **304 Not Modified**: The requested resource has not been modified since the last request.
- **307 Temporary Redirect**: The resource is temporarily located at a different URL, but the method should not change.
- **308 Permanent Redirect**: The resource is permanently located at a new URL, and the method should not change.

---

### **4xx: Client Errors**
These indicate errors caused by the client.
- **400 Bad Request**: The server could not understand the request due to invalid syntax.
- **401 Unauthorized**: The client must authenticate itself to get the requested resource.
- **402 Payment Required**: Reserved for future use; sometimes used for payment-related errors.
- **403 Forbidden**: The client does not have permission to access the resource.
- **404 Not Found**: The server cannot find the requested resource.
- **405 Method Not Allowed**: The request method is not supported for the requested resource.
- **406 Not Acceptable**: The requested resource is capable of generating only content not acceptable according to the Accept headers.
- **407 Proxy Authentication Required**: The client must authenticate with a proxy server before the request can be completed.
- **408 Request Timeout**: The server timed out waiting for the client’s request.
- **409 Conflict**: The request conflicts with the current state of the server.
- **410 Gone**: The resource is no longer available and will not be available again.
- **411 Length Required**: The server requires a Content-Length header in the request.
- **412 Precondition Failed**: The server does not meet one of the preconditions specified in the request headers.
- **413 Payload Too Large**: The request is larger than the server is willing or able to process.
- **414 URI Too Long**: The URI requested by the client is too long for the server to process.
- **415 Unsupported Media Type**: The server does not support the media format of the request.
- **416 Range Not Satisfiable**: The client requested a range of the resource that the server cannot supply.
- **417 Expectation Failed**: The server cannot meet the requirements of the Expect request-header field.
- **418 I’m a Teapot**: A playful code from the Hyper Text Coffee Pot Control Protocol.
- **429 Too Many Requests**: The client has sent too many requests in a given amount of time.

---

### **5xx: Server Errors**
These indicate that the server failed to fulfill a valid request.
- **500 Internal Server Error**: The server encountered an unexpected condition that prevented it from fulfilling the request.
- **501 Not Implemented**: The server does not support the functionality required to fulfill the request.
- **502 Bad Gateway**: The server, while acting as a gateway, received an invalid response from the upstream server.
- **503 Service Unavailable**: The server is currently unavailable (due to overload or maintenance).
- **504 Gateway Timeout**: The server, while acting as a gateway, did not receive a timely response from the upstream server.
- **505 HTTP Version Not Supported**: The server does not support the HTTP protocol version used in the request.
- **507 Insufficient Storage**: The server cannot store the representation needed to complete the request.
- **511 Network Authentication Required**: The client must authenticate to gain network access.

The `next_sibling` attribute is used in web scraping with BeautifulSoup to navigate between elements in the HTML tree. Specifically, it allows you to access the next element (or node) that appears at the same hierarchical level as the current one. 

Here's why and when you might use `next_sibling`:

### 1. **Accessing Text or Elements After a Specific Tag**
   If the text or data you want is not inside the tag you're currently targeting but is located immediately after it, `next_sibling` helps you navigate to that data.

   **Example HTML:**
   ```html
   <div>
       <span>Salary:</span>
       1 - 1 Lakhs
   </div>
   ```

   **Code Without `next_sibling`:**
   ```python
   span = soup.find('span', text='Salary:')
   print(span.text)  # Output: "Salary:"
   ```

   If you want to get the `1 - 1 Lakhs` part, you need to move to the next node in the tree.

   **Code With `next_sibling`:**
   ```python
   span = soup.find('span', text='Salary:')
   print(span.next_sibling.strip())  # Output: "1 - 1 Lakhs"
   ```

---

### 2. **When Data Is Outside the Target Tag**
   Sometimes, web pages are structured in a way where the target data is not nested inside the element you're searching for but is placed after it.

   **Example HTML:**
   ```html
   <i class="salary-icon"></i>
   1 - 1 Lakhs
   ```

   **Code:**
   ```python
   icon = soup.find('i', class_='salary-icon')
   print(icon.next_sibling.strip())  # Output: "1 - 1 Lakhs"
   ```

---

### 3. **Why Use `next_sibling` Instead of `get_text`?**
   - **`get_text()`** retrieves all the text inside a tag, including its children.
   - **`next_sibling`** is useful when the data is not inside the tag but adjacent to it.

---

### Common Issues with `next_sibling`:
1. **Whitespace Nodes:** Sometimes, `next_sibling` might return a newline or whitespace node instead of the expected data. Use `.strip()` to clean it or `find_next_sibling()` for more control.
2. **Nested Data:** If the data is deeply nested, you may need to chain multiple `.next_sibling` or use other methods like `find_next()`.

````python
# URL of the webpage
url = 'https://www.timesjobs.com/candidate/job-search.html?from=submit&luceneResultSize=25&txtKeywords=IT&postWeek=60&searchType=personalizedSearch&actualTxtKeywords=IT&searchBy=0&rdoOperator=OR&pDate=I&sequence=1&startPage=1'

# Send a request to the webpage
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')

# Extract job names
names = [name.text.strip() for name in soup.find_all('h2', class_='heading-trun')]

# Extract company names
company_names = [comp_name.text.strip() for comp_name in soup.find_all('h3', class_='joblist-comp-name')]

# Extract locations
locations = [location.text.strip() for location in soup.find_all('li', class_='srp-zindex location-tru')]

# Extract experience requirements
experiences = []
experience_elements = soup.find_all('i', class_='srp-icons experience')
for elem in experience_elements:
    experience = elem.next_sibling.strip().replace(' Years', '')
    experiences.append(experience)

# Create a DataFrame from the data collected
df = pd.DataFrame({
    'Job Title': names,
    'Company Name': company_names,
    'Location': locations,
    'Experience (Years)': experiences,
})

# Save the DataFrame to a CSV file 
df.to_csv('job_listings.csv', index=False)

# Display the DataFrame
print("Sample Job Listings")
print(df.head(10).to_string(index=False))
````

````python
final = pd.DataFrame()

# Loop through each page
for j in range(1, 11):  # Adjust the range as needed
    url = f'https://www.timesjobs.com/candidate/job-search.html?from=submit&luceneResultSize=25&txtKeywords=IT&postWeek=60&searchType=personalizedSearch&actualTxtKeywords=IT&searchBy=0&rdoOperator=OR&pDate=I&sequence={j}&startPage=1'
    
    # Send a request to the webpage
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    # Extract job names
    names = [name.text.strip() for name in soup.find_all('h2', class_='heading-trun')]
    
    # Extract company names
    company_names = [comp_name.text.strip() for comp_name in soup.find_all('h3', class_='joblist-comp-name')]
    
    # Extract locations
    locations = [location.text.strip() for location in soup.find_all('li', class_='srp-zindex location-tru')]
    
    # Extract experience requirements
    experiences = []
    experience_elements = soup.find_all('i', class_='srp-icons experience')
    for elem in experience_elements:
        experience = elem.next_sibling.strip().replace(' Years', '')
        experiences.append(experience)
    
    # Create a DataFrame from the data collected on this page
    df = pd.DataFrame({
        'Job Title': names,
        'Company Name': company_names,
        'Location': locations,
        'Experience (Years)': experiences,
    })
    
    # Append the DataFrame to the final result
    final = pd.concat([final, df], ignore_index=True)

# Save the combined DataFrame to a CSV file
final.to_csv('job_listings_all_pages.csv', index=False)

# Display a sample of the combined DataFrame
print("Sample Job Listings from All Pages")
print(final.head().to_string(index=False))
````
